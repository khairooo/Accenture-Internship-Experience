```python
# importing packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read csv files (REaction Types, Reactions, Content)
reaction_types = pd.read_csv("/content/ReactionTypes.csv")
reactions = pd.read_csv("/content/Reactions.csv")
content = pd.read_csv("/content/Content.csv")
```

## 2) Reactions

```python
reactions.head()
```

```python
# info checking
reactions.info()
```

- Depending on the informations We get, it is obvious that Datetime is object type. in the next steps We will transfer it to Datetime column,Where we will be able to analyse the dataset in different periods of time

```python
reactions.isnull().sum()
```

- Here We go, As We can see above that the reactions Dataset has missing values in two columns, UserID and Types,
- There are couple approaches to deal with missing values (replacing by mean median or mode, or any other technicsbut, in our task We are not obligated to use them, We will just remove them.

```python
# Drop missing values using dropna()
reactions = reactions.dropna()
```

```python
# checking that the missing values have been removed
reactions.isnull().sum()
```

- As showing above the all the rows with missing values do not exist again in reactions dataset

```python
# Datetime column : converting from Object type to Datetime
reactions['Datetime'] = pd.to_datetime(reactions['Datetime'])
reactions.info()
```

- Perfect, We have now converted Datetime column from Object to Datetime

## 3) Content

```python
content.head()
```

```python
content.isnull().sum()
```

URL column has missing values, by using the same function dropna(), We can remove all of them


```python
content = content.dropna()
```

```python
content.isnull().sum()
```

```python
# URL column removal
## Depending on the task URL,User ID  are not important columns
content = content.drop(['URL', 'User ID'], axis=1)
reactions = reactions.drop(['User ID'], axis=1)
```

```python
# let's check unique categories exists and try to clean more and more
content["Category"].value_counts()
```

- In the results above we get two different studying categories, Uppercase. Lowercase. Where we can merge them under the same category by lowercasing all the categories we have

```python
# Lowercase all the rows
content['Category'] = content['Category'].apply(lambda x: x.lower())
```

```python
content["Category"].value_counts()
```

```python
# rename type column to Content Type, and 'Unnamed: 0' to ID
content = content.rename(columns={"Type":"Content Type",'Unnamed: 0': 'ID'})
reaction_types = reaction_types.rename(columns={'Unnamed: 0': 'ID'})
reactions= reactions.rename(columns={'Unnamed: 0': 'ID'})

```

## Merging datasets

```python
merged_data = content.merge(reactions, on='Content ID').merge(reaction_types, on = "Type")
pd.DataFrame(merged_data)
```

```python
merged_data = merged_data.drop(["ID_x","ID_y","ID"],axis = 1)
merged_data
```

```python
# writing to Excel
datatoexcel = pd.ExcelWriter('merged_data.xlsx')

# write DataFrame to excel
merged_data.to_excel(datatoexcel)

# save the excel
datatoexcel.save()
print('DataFrame is written to Excel File successfully.')
```

```python
category_score = merged_data.groupby(["Category"])["Score"].sum().sort_values(ascending = False)
category_score
```

```python
# writing to Excel
datatoexcel = pd.ExcelWriter('category_score.xlsx')

# write DataFrame to excel
category_score.to_excel(datatoexcel)

# save the excel
datatoexcel.save()
print('category_score is written to Excel File successfully.')
```
